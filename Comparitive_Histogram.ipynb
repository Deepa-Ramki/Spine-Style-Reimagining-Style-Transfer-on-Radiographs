```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im1 = cv2.imread("AP.png") #C-arm image path
im2 = cv2.imread("DRR_AP.jpg") #DRR path
im3 = cv2.imread("op200.png") #SpineStyle
im4 = cv2.imread("op500.png") #VGG19

im5 = cv2.imread("vgg16OP.png") #VGG16
im6 = cv2.imread("AlexnetResult.png") #AlexNet
im7 = cv2.imread("inceptionOP.png") #Inceptionv3
im8 = cv2.imread("ResNet50_op_crop.jpg") #ResNet50
```

```python
plt.imshow(im1, cmap = 'grey')
```

```python
h3 = cv2.calcHist([im3],[0],None,[256],[10,250]).flatten()
image_data = im3.flatten()
hist, bins = np.histogram(image_data, bins=256)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
fig, ax = plt.subplots(figsize=(11, 4))
ax.plot(bin_centers, h1 ,linestyle = '--', lw = 2 , color="black", label = 'C-arm Image', alpha=1)
ax.fill_between(bin_centers, h3, color="blue", label = 'SpineStyle', alpha=0.6)
ax.fill_between(bin_centers, h2, color="red",label = 'DRR', alpha=0.5)
plt.title("Style Metric",fontsize=18 )
plt.xticks(fontsize = 15) 
plt.yticks(fontsize = 15) 
plt.xlabel('Pixel Intensity',fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

```python
h3 = cv2.calcHist([im3],[0],None,[256],[10,250]).flatten()
image_data = im3.flatten()
hist, bins = np.histogram(image_data, bins=256)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
fig, ax = plt.subplots(figsize=(18, 6))
ax.fill_between(bin_centers, h2, color="black", label = 'DRR', alpha=0.2)
ax.plot(bin_centers, h3, lw = 2, color="black", label = 'SpineStyle', alpha=1)
ax.plot(bin_centers, h4, lw = 2,color="green",label = 'VGG19', alpha=1)
ax.plot(bin_centers, h5, lw = 2 ,color="red",label = 'VGG16', alpha=1)
plt.title("Style Comparison - VGG family",fontsize=30 )
plt.xticks(fontsize = 25) 
plt.yticks(fontsize = 25) 
plt.xlabel('Pixel Intensity',fontsize=25)
plt.ylabel("Frequency",fontsize=25)
plt.legend(fontsize=28)
plt.show()
```
